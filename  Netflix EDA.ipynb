# Q5 Show only the Titles of all TV Shows that were released in India only.

# Q10 What is the maximum duration of a Movie/Show on Netflix?

```python
#To find the maximum duration of a show on netflix
df_duration = df.dropna()
df_new=df_duration[(df_duration['duration'].str.contains('Seasons'))]['duration'].unique()
```

```python
c=[]
for i in range(len(df_new)):
    c.append(df_new[i].replace('Seasons',''))

c_new = list(map(int, c))
c_new.sort(reverse=True)
max(c_new)
```

```python
df[(df['duration']=='15 Seasons')]
```

```python
df_new_2=df_duration[(df_duration['duration'].str.contains('min'))]['duration'].unique()
```

```python
k=[]
for i in range(len(df_new_2)):
    k.append(df_new_2[i].replace('min',''))
    
k_new = list(map(int,k))
k_new.sort(reverse=True)
print(max(k_new))
```
